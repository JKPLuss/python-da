### EXERCISE 9

Illustarate solution of 2-3 systems of linear equations from the "Mathematics for Data Analytics" course using np.linalg

```python
import numpy as np

A = np.array([
    [ 3, -2,  1,  4],
    [ 1,  0,  2, -1],
    [ 5,  1, -1,  2],
    [ 2,  3,  4,  1]
], dtype=float)
b = np.array([
    [15],
    [ 7],
    [ 8],
    [14]
    ], dtype=float)
x = np.linalg.solve(A, b)
print(f"x1={x[0]}, x2={x[1]}, x3={x[2]}, x4={x[3]}" )
A_inv = np.linalg.inv(A)
print(f"Inverse matrix =\n{A_inv}")
print("Transpose A.T:\n", A.T)
det = np.linalg.det(A)
print("Determinant:", det)
inv_solv = np.dot(A_inv,b)
print(f"x1={inv_solv[0]}, x2={inv_solv[1]}, x3={inv_solv[2]}, x4={inv_solv[3]}")
#Rank
Ab = np.hstack((A, b))
rankA = np.linalg.matrix_rank(A)
rankAb = np.linalg.matrix_rank(Ab)
n = A.shape[1]
print("rank(A)  =", rankA)
print("rank([A|b]) =", rankAb)
print("variables =", n)
if rankA == rankAb == n:
    print("One solution")
elif rankA == rankAb < n:
    print("Infinitinve solution")
elif rankA < rankAb:
    print("No solution")

```

### EXERCISE 10

Load numeric data from the CSV file, utilised in Lab 3 Exercise 9 into a NumPy array and calculate its statistics (by columns)

```python
import csv
import random
import json
import numpy as np

data = np.genfromtxt(
    '/content/sample_data/ispd_pasvaldibas_dzimumi_20250701.csv',
    delimiter=',',
    dtype=None,
    names=True,
    encoding='utf-8'
)
col_names = data.dtype.names

numeric_cols = []
for name in col_names:
    if np.issubdtype(data[name].dtype, np.number):
        numeric_cols.append(name)
numeric = np.column_stack([data[name] for name in numeric_cols])
print(f"columns = {len(numeric_cols)}")
mean_vals = np.mean(numeric, axis=0)
median_vals = np.median(numeric, axis=0)
std_vals = np.std(numeric, axis=0)
var_vals = np.var(numeric, axis=0)
min_vals = np.min(numeric, axis=0)
max_vals = np.max(numeric, axis=0)
print("mean =", [f"{v:.2f}" for v in mean_vals])
print("median =", [f"{v:.2f}" for v in median_vals])
print("std =", [f"{v:.2f}" for v in std_vals])
print("var =", [f"{v:.2f}" for v in var_vals])
print("min =", [f"{v:.2f}" for v in min_vals])
print("max =", [f"{v:.2f}" for v in max_vals])
```
