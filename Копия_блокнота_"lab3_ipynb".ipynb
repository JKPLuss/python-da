### EXERCISE 1

Write a function _info_ that returns information about the argument:

* type
* size in the memory
* length (for sequence objects)
* hash (if hashable)

Test the function for the following types of objects: bool, str, int, float, list, tuple, dict, set

```python
import sys

def info(obj):
  if isinstance(obj, (bool, int, float)):
    return f"{type(obj)}, size= {sys.getsizeof(obj)}, Length = No Length, hash= {hash(obj)}"
  elif isinstance(obj, (list,dict,set)):
    return f"{type(obj)}, size= {sys.getsizeof(obj)}, Length = {len(obj)}, hash= unhashable type"
  return f"{type(obj)}, size= {sys.getsizeof(obj)}, Length = {len(obj)}, hash= {hash(obj)}"
print(info(True))
print(info("flor"))
print(info(1))
print(info(1.2))
print(info([1,2,3]))
print(info((1,2,3)))
print(info({"a":1,"b":2}))
print(info({1,2,3}))
```

### EXERCISE 3

Write a programme that returns a list of first _n_ powers of 2:

get_list_2(5)

returns

2, 4, 8, 16, 32


```python
def get_list_2(n):
  result = []
  for i in range(n):
    result.append(2**(i+1))
  return result
print(get_list_2(5))
```

### EXERCISE 4

Write a programme that returns a list of powers of 2 between two specified arguments:

get_list_b(5, 50)

returns

8, 16, 32

```python
def get_list_b(x1,x2):
  result=[]
  n=1
  while True:
    if 2**n>x2:
      break
    if 2**n>=x1:
      result.append(2**n)
    n+=1
  return result
print(get_list_b(5,50))
```

### EXERCISE 6

Write a function that returns the most frequently used word in the text argument

```python
def most_used_words(filename, top=10):
  words = dict()
  with open(filename, 'r',encoding='utf-8',) as file:
    text = file.read()
    word_list = text.split()
    for word in word_list:
      if word.lower() in words:
        words[word.lower()]+=1
      else:
        words[word.lower()]=1
  result = sorted(words.items(), key=lambda x: x[1], reverse=True)
  if len(result)<top:
    return result
  return result[:top]


most_used_words("/content/sample_data/alice_in_wonderland.txt")
```

### EXERCISE 7

Write a function that use a list argument and returns a dictionary, where every list element is linked with its number of occurances in the list:

todict(["a", "b", "c", "b", "b", "c"])

returns

{
    "a": 1,
    "b": 3,
    "c": 2
}


```python
def list_to_dict(work_list):
  result = dict()
  for item in work_list:
    if item in result:
      result[item]+=1
    else:
      result[item]=1
  return result
list_to_dict(["a", "b", "c", "b", "b", "c"])
```

### EXERCISE 11

Develop a function that saves Python objects into JSON file of the given name and another function to restore them.

Illustrate usage of these functions for saving/restoring data structures from Lab 2 Exercise 10, Exercise 11, and Exercise 12.

```python
import json

def save_json(filename, obj):
  with open(filename, 'w', encoding='utf-8') as file:
    json.dump(obj, file, ensure_ascii=False, indent=2)
def load_json(filename):
    with open(filename, "r", encoding="utf-8") as file:
        return json.load(file)

data = {
    "name": "Jura",
    "roles": ["owner", "developer"],
    "age": 45,
    "active": True
}

save_json("/content/sample_data/user_data.json",data)

restored = load_json("/content/sample_data/user_data.json")
print(json.dumps(restored, indent=2))
```
